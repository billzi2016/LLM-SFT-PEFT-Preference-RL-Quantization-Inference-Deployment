# 常见错误

这个 notebook 汇总训练、量化、部署中最常见的理解错误和排查方向。

## 1. QLoRA 误区

错误理解：QLoRA 会训练出一个 4bit adapter。

正确理解：QLoRA 是 4bit 加载基座模型，然后训练 LoRA adapter。adapter 本身通常不是所谓 4bit 完整模型。

## 2. LoRA adapter 误区

错误理解：adapter 可以单独部署。

正确理解：adapter 是增量参数。推理时要么和基座模型一起加载，要么先 merge 到基座模型。

## 3. 长上下文误区

错误理解：模型支持 128K，就应该默认开 128K。

正确理解：上下文越长，KV cache 和训练显存越高。真实部署要根据显存、吞吐和任务需求设置。

## 4. 量化格式误区

错误理解：GGUF、AWQ、GPTQ、EXL2、FP8、MXFP4 都是差不多的 4bit。

正确理解：

- GGUF 是本地推理格式。
- AWQ/GPTQ/EXL2 是离线推理量化路线。
- FP8 是新硬件服务端推理低精度路线。
- MXFP4 是 gpt-oss 相关重点压缩格式。
- bitsandbytes 4bit 是训练/加载路线。

## 5. 部署路线误区

错误理解：Transformers、vLLM、Ollama、SGLang 是同一种东西。

正确理解：

- Transformers 用于最小验证。
- vLLM 用于服务端高吞吐 API。
- Ollama 用于本地简单运行。
- SGLang 用于复杂生成控制。
- llama.cpp 用于 GGUF 和低资源场景。

## 6. gpt-oss 误区

错误理解：gpt-oss 可以按普通 ChatML 模型处理。

正确理解：gpt-oss 使用 Harmony 格式，并且有 reasoning effort、MoE、MXFP4 等额外注意点。模板错误会导致输出边界和最终回答不稳定。
