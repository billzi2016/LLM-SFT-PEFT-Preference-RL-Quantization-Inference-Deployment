# AWQ、GPTQ、EXL2 量化

这个 notebook 集中说明三类离线推理量化：AWQ、GPTQ、EXL2。

它们主要用于推理，不是训练起点。训练通常先得到基座模型或合并后的模型，再做这些量化。

## 1. 对比

| 方法 | 生态 | 典型用途 |
| --- | --- | --- |
| AWQ | AutoAWQ、vLLM、Transformers | 服务端 4bit 推理 |
| GPTQ | AutoGPTQ、Optimum、Transformers | GPU 推理量化 |
| EXL2 | ExLlamaV2 | 单机 GPU 高速推理 |
